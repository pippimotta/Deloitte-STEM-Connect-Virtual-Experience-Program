```python
import pandas as pd
```

```python
df = pd.read_excel('Task 5 Equality Table.xlsx')
df.head()
```

```python
def equalityCheck(score):
  if score <-20 or score > 20:
    return 'Highly Discriminative'
  elif score < -10 or score >10:
    return 'Unfair'
  else:
    return 'Fair'

df['Equality class'] = df['Equality Score'].apply(equalityCheck)
```
